```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, classification_report ,mean_squared_error,mean_absolute_error


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
data = data.drop(['file','urban_imperial','extra_urban_imperial','combined_imperial',
                  'tax_band','thc_emissions','particulates_emissions','first_year_12_months',
                  'first_year_6_months','date_of_change','standard_6_months','standard_12_months',
                  'thc_nox_emissions','nox_emissions','fuel_cost_6000_miles','transmission',
                  'fuel_cost_12000_miles','description','model','manufacturer'],axis=1)


mean_urban_metric = data['urban_metric'].mean() 
#print(mean_urban_metric)

mean_extra_urban_metric = data['extra_urban_metric'].mean()
#print(mean_extra_urban_metric)

mean_combined_metric = data['combined_metric'].mean()

mean_noise = data['noise_level'].mean()

mean_eng_cap = data['engine_capacity'].mean() 

mean_co_emissions = data['co_emissions'].mean()

mean_co_emissions = data['co_emissions'].mean()


data['engine_capacity'] = data['engine_capacity'].fillna(value = mean_eng_cap)
data['noise_level'] = data['noise_level'].fillna(value = mean_noise)
data['combined_metric'] = data['combined_metric'].fillna(value = mean_combined_metric)
data['extra_urban_metric'] = data['extra_urban_metric'].fillna(value = mean_extra_urban_metric)
data['urban_metric'] = data['urban_metric'].fillna(value = mean_urban_metric)
data['co_emissions'] = data['co_emissions'].fillna(value = mean_co_emissions)

data['transmission_type'] = data['transmission_type'].fillna(value = 'Manual')

data['transmission_type'] = np.where(data['transmission_type']=='Automatic',1 ,data['transmission_type'])
data['transmission_type'] = np.where(data['transmission_type']=='Manual',0 ,data['transmission_type'])

data['transmission_type'] = data['transmission_type'].astype("int64")


transmission_type_dummies = pd.get_dummies(data['transmission_type'], drop_first=True)
data = data.drop(['transmission_type'],axis=1)
data = pd.concat([data,transmission_type_dummies],axis=1)

#print(transmission_type_dummies)
#print(data['transmission_type'].head(30))

fuel_dummies = pd.get_dummies(data['fuel_type'], drop_first=True)
data = data.drop(['fuel_type'],axis=1)
data = pd.concat([data,fuel_dummies],axis=1)
```

```python
X = data.drop('co2',axis=1).values
y = data['co2'].values

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=101)

MMS_scaler = MinMaxScaler()

X_train = MMS_scaler.fit_transform(X_train)

X_test = MMS_scaler.transform(X_test)

Seq_model = Sequential()

Seq_model.add(Dense(120,activation = 'relu'))
Seq_model.add(Dense(120,activation = 'relu'))
Seq_model.add(Dense(120,activation = 'relu'))
Seq_model.add(Dense(120,activation = 'relu'))
Seq_model.add(Dense(1))

Seq_model.compile(optimizer = 'adam' , loss = 'mse')

Seq_model.fit(X_train,y_train, validation_data = (X_test,y_test) , batch_size = 128 , epochs = 250)

predictions = Seq_model.predict(X_test)

print("MSE : ", mean_squared_error(predictions,y_test))
print("\n","MAE : ", mean_absolute_error(predictions,y_test))
```
